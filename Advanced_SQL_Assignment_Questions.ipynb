**Q1. What is a Common Table Expression (CTE), and how does it improve SQL query readability?**
- A Common Table Expression (CTE) is a temporary, named result set that you define within a SQL query using the WITH keyword. It exists only for the duration of that query and can be referenced just like a table.

Basic Syntax

WITH cte_name AS (

    SELECT column1, column2
    FROM table_name
    WHERE condition
)
SELECT *
FROM cte_name;

Simple Example

WITH HighSalary AS (

    SELECT Name, Salary
    FROM Employees
    WHERE Salary > 50000
)
SELECT *
FROM HighSalary;

**Q2. Why are some views updatable while others are read-only? Explain with an example.**
- Views are updatable when they are simple and based on a single table.
They become read-only when they involve joins, grouping, or complex queries because SQL cannot determine how to update the original data.

Example

- Create view
CREATE VIEW emp_view AS
SELECT employee_id, first_name, salary
FROM employees;

- Update through view
UPDATE emp_view
SET salary = 60000
WHERE employee_id = 101;

**Q3. What advantages do stored procedures offer compared to writing raw SQL queries repeatedly?**
- Stored procedures improve performance, security, and reusability by allowing pre-written SQL code to be stored and executed multiple times instead of writing the same queries repeatedly.

Example

- Create Stored Procedure
CREATE PROCEDURE GetITEmployees(

)
BEGIN

    SELECT *
    FROM employees
    WHERE department = 'IT';

END;

- Call Stored Procedure
CALL GetITEmployees();

**Q4. What is the purpose of triggers in a database? Mention one use case where a trigger is essential.**
- A Trigger is a special type of SQL program that automatically executes when a specific event occurs in a table.

These events include:

INSERT (adding data)

UPDATE (modifying data)

DELETE (removing data)

- Purpose of Triggers

To automate actions without manual intervention

To maintain data consistency and integrity

To enforce business rules

To track or log changes in the database

**Use Case (Important)**

Audit / Tracking Changes

In companies, it is important to track salary updates.

A trigger ensures that every change is automatically recorded without relying on manual entry.

**Q6. Write a CTE to calculate the total revenue for each product
 (Revenues = Price × Quantity), and return only products where  revenue > 3000.**
 - A CTE is used to calculate revenue per product, and only those products with revenue greater than 3000 are selected.

 - Create CTE to calculate total revenue
WITH RevenueCTE AS ( SELECT
        p.ProductID,
        p.ProductName,
        (p.Price * s.Quantity) AS Revenue

    FROM Products p
    JOIN Sales s
    ON p.ProductID = s.ProductID
)

-  Display products with revenue greater than 3000
SELECT *
FROM RevenueCTE
WHERE Revenue > 3000;

**Q7. Create a view named vw_CategorySummary that shows:
Category, TotalProducts, AveragePrice.**
- A view is created to summarize product data by category, showing total products and average price for each category.

- Create view to show category-wise summary

CREATE VIEW vw_CategorySummary AS
SELECT

    Category,
    COUNT(ProductID) AS TotalProducts,
    AVG(Price) AS AveragePrice

FROM Products
GROUP BY Category;

category**

**Q10. Create an AFTER DELETE trigger on the Products table that archives deleted product rows into a new table ProductArchive. The archive should store ProductID, ProductName, Category, Price, and DeletedAt timestamp.**

- To create an AFTER DELETE trigger that archives product data, you can use the following SQL syntax. This script assumes the ProductArchive table already exists with the specified columns.

DELIMITER //

CREATE TRIGGER AfterProductDelete
AFTER DELETE ON Products
FOR EACH ROW
BEGIN

    INSERT INTO ProductArchive (ProductID, ProductName, Category, Price, DeletedAt)
    VALUES (OLD.ProductID, OLD.ProductName, OLD.Category, OLD.Price, NOW());
END //

DELIMITER ;

When a product is deleted, the trigger automatically saves its data into ProductArchive.
